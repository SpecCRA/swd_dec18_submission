# Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Load files

```python
# On Mini
origins_file = None
dest_file = None

#on Macbook
origins_file = 'airport_origins.csv'
dest_file = 'airport_dest.csv'

odf = pd.read_csv(origins_file) # Load origins data
ddf = pd.read_csv(dest_file) # Load destinations data
```

```python
# Drop odd column in DataFrame after loading
odf = odf.drop(columns = 'Unnamed: 0')
ddf = ddf.drop(columns = 'Unnamed: 0')
```

```python
# Rename city columns for easier access
odf = odf.rename(columns = {'Origin City': 'City'})
ddf = ddf.rename(columns = {'Destination City': 'City'})

# Grab the data I need
odf = odf[['City', 'Year', 'Rank']]
ddf = ddf[['City', 'Year', 'Rank']]
```

```python
cities = odf.City.unique()

for city in cities:
    x = odf[odf.City == city].Year
    y = odf[odf.City == city].Rank
    plt.plot(x,y, '-o', label = city, markersize = 30)
    #plt.plot(x,y)

plt.rcParams['figure.figsize'] = [50,30]
plt.ylim(11, 0)
plt.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.2, fontsize = 25)
plt.xticks(range(2000, 2010), fontsize = 35)
plt.yticks(range(1,11), fontsize = 25)
plt.title('Chicago has the Busiest Domestic Airport from 2000 to 2009', fontsize = 60, loc = 'left')
plt.xlabel('Year', fontsize = 50, horizontalalignment = 'right')
plt.ylabel('Rank of Total Domestic Flights', fontsize = 50)
#plt.show()
plt.savefig('busy_airports.png')
```
